```python
churn_rate = df['Churn'].value_counts(normalize=True) * 100
print("Taxa de Churn (%): {}".format(churn_rate))
```

```python

churn_data = pd.Series(
    [71.198569, 25.719004, 3.082427],
    index=['NÃ£o Churn', 'Churn', 'Desconhecido'],
    name='Taxa de Churn (%)'
)

colors = ['#5cb85c', '#d9534f', '#f0ad4e']


churn_data.plot(
    kind='pie',
    autopct='%.1f%%',
    startangle=90,
    colors=colors,
    ylabel='',
    title='Taxa de Churn'
)

plt.axis('equal')  # Para manter formato de cÃ­rculo
plt.tight_layout()
plt.savefig('data/churn_rate.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
churn_genero = dados.groupby('customer.gender')['Churn'].value_counts(normalize=True).unstack() * 100
churn_genero
```

```python
ax = churn_genero.plot(kind='bar', color=['#5cb85c', '#d9534f'])
plt.title('Taxa de Churn por GÃªnero')
plt.xlabel('GÃªnero')
plt.ylabel('Porcentagem (%)')
plt.legend(title='Churn', labels=['NÃ£o Churn', 'Churn'])
plt.xticks(ticks=[0, 1], labels=['Masculino', 'Feminino'], rotation=0)
plt.xticks(rotation=0)

for p in ax.patches:
    ax.annotate(f"{p.get_height():.1f}%",
                (p.get_x() + p.get_width() / 2, p.get_height() + 1),
                ha='center', va='bottom', fontsize=10, color='black')

plt.ylim(0, 100)
plt.tight_layout()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.savefig('data/churn_genero.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
churn_idade = dados.groupby('customer.SeniorCitizen')['Churn'].value_counts(normalize=True).unstack() * 100
churn_idade
```

```python
ax = churn_idade.plot(kind='bar', color=['#5cb85c', '#d9534f'])
plt.title('Taxa de Churn por Idade')
plt.xlabel('Idade')
plt.ylabel('Porcentagem (%)')
plt.xticks(ticks=[0, 1], labels=['NÃ£o Idoso', 'Idoso'], rotation=0)
for p in ax.patches:
    ax.annotate(f"{p.get_height():.1f}%",
                (p.get_x() + p.get_width() / 2, p.get_height() + 1),
                ha='center', va='bottom', fontsize=10, color='black')
plt.legend(title='Churn', labels=['NÃ£o Churn', 'Churn'], loc='upper right')
plt.ylim(0, 100)
plt.tight_layout()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.savefig('data/churn_idade.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
churn_tipo_contrato = dados.groupby('account.Contract')['Churn'].value_counts(normalize=True).unstack() * 100
churn_tipo_contrato
```

```python
ax = churn_tipo_contrato.plot(kind='bar', color=['#5cb85c', '#d9534f'])
plt.title('Taxa de Churn por Tipo de Contrato')
plt.xlabel('Tipo de Contrato')
plt.ylabel('Porcentagem (%)')
plt.xticks(
    ticks=range(len(churn_tipo_contrato.index)),
    labels=['MÃªs a MÃªs', 'Um Ano', 'Dois Anos'],
    rotation=0
)
for p in ax.patches:
    ax.annotate(f"{p.get_height():.1f}%",
                (p.get_x() + p.get_width() / 2, p.get_height() + 1),
                ha='center', va='bottom', fontsize=10, color='black')
plt.legend(title='Churn', labels=['NÃ£o Churn', 'Churn'], loc='upper right')
plt.ylim(0, 100)
plt.tight_layout()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.savefig('data/churn_tipo_contrato.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
mapeamento = {
    'Electronic check': 'Cheque EletrÃ´nico',
    'Mailed check': 'Cheque Enviado',
    'Bank transfer (automatic)': 'TransferÃªncia BancÃ¡ria (AutomÃ¡tica)',
    'Credit card (automatic)': 'CartÃ£o de CrÃ©dito (AutomÃ¡tico)'
}

churn_tipo_pagamento = dados.groupby('account.PaymentMethod')['Churn'].value_counts(normalize=True).unstack() * 100
churn_tipo_pagamento.rename(index=mapeamento, inplace=True)
churn_tipo_pagamento.sort_values(by='Yes', ascending=True, inplace=True)
churn_tipo_pagamento = churn_tipo_pagamento.rename(
    columns={'No': 'NÃ£o Churn', 'Yes': 'Churn'}
)
churn_tipo_pagamento
```

```python
total_churn_tipo_pagamento = dados.groupby('account.PaymentMethod')['Churn'].value_counts().unstack()
total_churn_tipo_pagamento
```

```python
ax = churn_tipo_pagamento.plot(kind='barh', color=['#5cb85c', '#d9534f'])
plt.title('Taxa de Churn por Tipo de Pagamento')
plt.xlabel('Porcentagem (%)')
plt.ylabel('Tipo de Pagamento')

for p in ax.patches:
    width = p.get_width()
    y_center = p.get_y() + p.get_height() / 2
    ax.annotate(f"{width:.1f}%",
                (width + 1, y_center),
                ha='left', va='center', fontsize=7, color='black')

plt.legend(title='Churn', labels=['NÃ£o Churn', 'Churn'], loc='upper right')
plt.xlim(0, 100)
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.savefig('data/churn_tipo_pagamento.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
churn_tenure = dados.groupby('customer.tenure')['Churn'].value_counts(normalize=True).unstack() * 100
churn_tenure = churn_tenure[churn_tenure.index <= 12]
churn_tenure
```
